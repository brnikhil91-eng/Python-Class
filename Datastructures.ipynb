LISTS -
Heterogeneous, mutable(CAN BE CHANGED), orderd, duplicates are allowed.
It is stored in [] brackets.

```python
a=[1,3, "apple", 4.5, True, 4+5j, "kiwi"]
print(a)
print(a[3]) #INDEXING
print(a[0:4]) #SLICING

a[4]="Orange"#replacing text
print(a)
```

LIST METHODS:
Append, remove, extend, insert, pop, clear, index, count, sort, reverse, copy, deepcopy

```python
a=[1,2,3,4] #append - adds any elements in the list and sits at the end of the list.
a.append(5)
print(a)

b=9
a.append(b)
print(a)

b=[10,11]
a.append(b)
print(a)
```

```python
#Extend - to add multiple elements into the list, use square bracket to add multiple elements into the list 

a=[1,2,3,4,5]
a.extend([6,5,6])
print(a)

b=[9,10,11]
a.extend(b)
print(a)

b=[9,10,11]
b.extend(a)
print(b)
```

```python
#INDEX - gives the position of the element starting from 0,1,2,3,4
a=[1,2,3,4,5]
a.index(3)
```

```python
#INSERT - inserting a value, (index position element to insert)
a.insert(0,1)
print(a)
```

```python
#Pop - removes the last element 
a.pop()
print(a)

a.pop(0)
print(a)
```

```python
#Remove - removes a specific element and not the index positioning
a=[1,2,3,4,5]
a.remove(4)
print(a)

b=[5,6,9,1,2,3,7]
b.remove(5)
print(b)
```

```python
#Clear
a.clear()
print(a)
```

```python
#Count - shows how many it has been repeated in the list 
a=[1,2,3,4,4,7,8,9,9,9,10,11,34]
a.count(9)
```

```python
#Sort

a.sort() #ascending
print(a)

a.sort(reverse=True) #descending
print(a)

a.reverse()
print(a)
```

```python
#Copy - can be used for simple elements and do not have any complex elememts
a=[1,2,3,4]
b=a.copy()
print(b)

a.append(5)
print(a)
print(b)
```

```python
a=[1,2,3,4]
b=a

print(a)
print(b)

a.append(5)
print(a)
print(b)
```

```python
#Deepcopy - copies all the elements in the list including list inside a list
import copy
a=[1,2,3,4,[5,6]]
b=copy.deepcopy(a)
print(b)

b[4][0]=100 #4th position and 0th element to replace
print(b)
```

TUPLES-
tuples are orderd, heteregeneous, (immutable)unchangable, allows duplicates
It is stored in () brackets

```python
a=(1,2,3,"red",1,5,4,2,3)
print(type(a))
```

```python
print(a[0])
print(a[0:3])
```

```python
#Tuple methods are count and index

a=(3,5,7,6,6,3,5,4)
print(a.count(3))
print(a.index(7))

t=6,8,7   #tuple packing
print(type(t))

a,b,c=t   #tuple unpacking
print(a)
print(b)
print(c)
```

```python
#Built-in functions that can be used in tuples:
t=(3,4,6,7,8,3)
print(len(t))
print(min(t))
print(max(t))
print(sum(t))
```

SETS-
Items are Unordered, heterogenous duplicate are not allowed, elements must be unique and immutable and can contain any type of elements such as integer, float, tuple.etc
It is stored in {} brackets

HASHABILITY:
ELEMENTS STORED WITHIN THE SET MUST BE IMMMUTABLE WHICH MEANS HASHABILITY IS FOLLOWED
AN ELEMENT IS CALLED HASHABLE WHEN IT IS OF IMMUTABLE TYPE & CAN BE STORED WITHIN SET.

```python
#Defining a set
a={"Monday", 2,4,2.5,True,4+2j,2,4} #No duplicates are allowed and the execution is unordered
print(a)
print(type(a))

b=set(["tuesday",2,2.5,False,3+4j]) #Set method 
print(type(b))
```

```python
#Removing duplicates using set method
a=[1,2,2,1,4,5,"mONDAY","Monday"]
print(a)

b=set(a)
print(b)
```

SET OPERATIONS-
UNION
INTERSECTION
DIFFERENCE
SYMMETRIC DIFFERENCE

```python
a={12,13,14}
b={14,15,16}

print(a|b) #Union of 2 sets
print(a & b) #Intersection of 2 sets
print(a-b) #Difference (in a but not in b)
print(a^b) #Symmetric difference (elements that are not common)
```

SET METHODS-
ADD
UPDATE
REMOVE
DISCARD
POP
CLEAR
ISSUBSET
ISSUPERSET
ISDISJOINT
COPY

```python
#ADD
s={1,2,3} #individual elements
s.add(4)
print(s)

#UPDATE #multiple elements
s={1,2,3}
s.update([2,4,9,5]) #list
s.update({7,8,5}) #set 
s.update((10,11,12)) #tuple
print(s)

#REMOVE
s.remove(5)
print(s)
s.discard(15) #also removes values even if the value is not present it executes
print(s)

#POP
s.pop()
print(s)

#CLEAR #clears all
s.clear()
print(s)

#ISSUBSET, ISSUPERSET, ISDISJOINT
s={1,2,3}
t={1,2,3,4,5,6}
u={4,5,6}
print(s.issubset(t))
print(s.issuperset(s))
print(s.isdisjoint(u)) #checks for no common elements

#COPY
b=s.copy()
print(b)
```

UPDATE OPERATIONS-
DIFFERENCE_UPDATE
INTERSECTION_UPDATE
SYMMETRIC_DIFFERENCE_UPDATE

```python
a={1,2,3,4,5}
b={2,3}

a.difference_update(b) #removes the diffenence
print(a)
```

```python
a={1,2,3,4,5}
b={2,3}
a.intersection_update(b) #keeps only the common
print(a)
```

```python
a={1,2,3}
b={2,3,4}
a.symmetric_difference_update(b) #removes common elements in both the sets
print(a)
```
