```python
import pandas as pd
import numpy as np

data={
    'Customer_name':["Asha","Bala","Chitra","Deepa","Esha","Bala","Bala","Bala"],
    'City':['Bangalore','Delhi','Mumbai','Delhi','Bangalore','Delhi','Delhi','Delhi'],
    'Order_value':[5000,7000,np.nan,6500,7200,7000,7000,7000],
    'Email':['asha123@gmail.com','bala@gmail.com','chitra@company.com','deepa_98@yahoo.com','esha@gmail.com','bala@gmail.com','bala@gmail.com','bala@gmail.com']
}

df=pd.DataFrame(data)
df
```

HANDLING DUPLICATES VALUES:-
DUPLICATES OCCUR WHEN THE SAME RECORD APPEARS MULTIPLE TIMES IN A DATASET.

COMMON REASONS:
- DATA ENTRY ERRORS
- MERGING DATASETS
- SYSTEM LOGGING ERRORS

DUPLICATES CAN DISTORT THE ANALYSIS, FOR EXAMPLE:
REVENUE MAY APPEAR HIGHER THEN REALITY  

```python
#Identifying duplicates
df.duplicated()
```

```python
#Viewing duplicate rows
df[df.duplicated()]
```

```python
#Removes completely identical rows
df.drop_duplicates()
```

```python
#using subset

df.drop_duplicates(subset='Email')
#this removes duplicates only based on the Email column
```

```python
#Keep decides with duplicate record remains
df.drop_duplicates(keep='first')
#keeps the first duplicate record
```

```python
df.drop_duplicates(keep='last')
#Keeps the last duplicate record
```

```python
df.drop_duplicates(keep=False)
#Keep parameter accepts only 3 values- first, last, false
```

Regex - Regular Expressions

TO EXTRACT PATTERNS FROM THE TEXT

COMMON REGEX PATTERNS-

\d -  DIGIT
\w - LETTER OR NUMBER
^  - START OF A STRING
$  - END OF A STRING
.  - ANY CHARACTER
+  - ONE OR MORE
*  - ZERO OR MORE


COMMONLY USED WITH -

.str.constains()
.str.extract()
.str.replace()
.str.match()

```python
import pandas as pd

data={"Email":["Asha123@gmail.com",
              "Bala@gmail.com",
              "Chitra@company.com",
              "Deepa_98@yahoo.com"]}

df=pd.DataFrame(data)
```

```python
# extract numbers from text
df["Email"].str.extract(r"(\d+)")
```

```python
#FIND ROWS CONTAINING NUMBERS
df["Email"].str.contains(r"\d")
```

```python
#EXTRACT EMAIL DOMAIN
df["Email"].str.extract(r"@(\w+)")
```

WHEN WE WRITE @(\w+) :-

() - CAPTURE GROUP
\w - LETTER OR NUMBER
+  -  ONE OR MORE

HERE MATCH STARTS AT THE SYMBOL @ BUT SINCE @ IS OUTSIDE OF THE CAPTURE GROUP IT IS NOT EXTRACTED
SO ONLY WHAT IS INSIDE THE CAPTURE GROUP() IS EXTRACTED.

```python
#FIND EMAILS ENDING WITH gmail.com
df["Email"].str.contains(r"@gmail\.com$")
```

```python
data={"Name":["Asha","Bala","Charlie","Deepak"]}
df=pd.DataFrame(data)
```

```python
#FIND NAMES STARTING WITH A
df["Name"].str.contains(r"^A")
```

```python
#REMOVE NUMBERS FROM EMAILS
df["Email"].str.replace(r"\d","",regex=True)
```

```python
#REGEX GROUPS
#Deepa_98@gmail.com
#WE WANT TO EXTRACT BOTH NAME & NUMBER IN THE EMIAL SEPARATELY

df["Email"].str.extract(r"(\w+)_(\d+)")
```
