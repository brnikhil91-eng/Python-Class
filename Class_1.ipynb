f-string formatting types: 

variable {x}
expression {x+y}
precision {x:2f}
width {x:8}
alignment {x:^8}
padding {x:*^8}
zero padding {x:05}
percentage {x:2%}
commas {x:,}

```python
print(f"The sum of the numbers is {a+b}") #expression

price=9606.5628
print(f"{price:.2f}") #precision

x=45
print(f"{x:9}") # width, gives spaces and then the value in the 9th place

p="python"
print(f"|{p:<15}|") #left alignment
print(f"|{p:>15}|") #right alignment
print(f"|{p:^15}|") #center alignment

print(f"{45:*<6}") #Padding (left, right and center)
print(f"{45:*>6}")
print(f"{45:*^6}")

print(f"{7:03}") #zero padding
print(f"{125:09}")

value=3.5
print(f"{value:07.2f}") #zero padding + percision

ratio=0.7862
print(f"{ratio:.3%}") #percentage formatting

salary=2259000000
print(f"{salary:,}")  #commas(thousands separator)
print(f"{salary:,.2f}")
```

| *Operator Type*                                          | *Associativity*   | *Example*   | *Explanation*                    |
| ---------------------------------------------------------| ----------------- | ------------| -------------------------------- |
| +, -, *, /, //, %, and, or, ==, > etc.                   | **Left to Right** | 10 - 4 + 2  | Evaluated as (10 - 4) + 2        |
| =, +=, -=, /=, etc.                                      | *Right to Left*   | x = y = 10  | Assigns 10 to y, then x          |
| ** (Exponentiation)                                      | **Right to Left** | 2 ** 3 ** 2 | Evaluated as 2 ** (3 ** 2) = 512 |
| Unary operators (+x, -x, ~x)                             | *Right to Left*   | -~5         | -(-6) = 6                        |

| *Precedence Level*   | *Operators*                                           | *Example*       |             |    |
| -------------------- | ------------------------------------------------------| ----------------| ----------- | -- |
| 1                    | () Parentheses                                        | (3 + 2) * 4     |             |    |
| 2                    | ** Exponentiation                                     | 2 ** 3 ** 2     |             |    |
| 3                    | +x, -x, ~x Unary plus, minus, bitwise NOT             | -5, ~2          |             |    |
| 4                    | *, /, //, %                                           | 10 / 2 * 3      |             |    |
| 5                    | +, - (Addition, Subtraction)                          | 5 + 3 - 2       |             |    |
| 6                    | <<, >> (Bitwise shift)                                | 4 << 1          |             |    |
| 7                    | & Bitwise AND                                         | a & b           |             |    |
| 8                    | ^, `                                                  | Bitwise XOR, OR | a ^ b, a| b |
| 9                    | Comparison operators (<, >, <=, >=, !=, ==)           | a > b           |             |    |
| 10                   | not                                                   | not True        |             |    |
| 11                   | and                                                   | a and b         |             |    |
| 12                   | or                                                    | a or b          |             |    |
| 13                   | Assignment (=, +=, -= etc.)                           | x += 2          |             |    |
| 14                   | , (comma)                                             | a, b = 1, 2     |             |    |

| *Aspect*          | *Operator Precedence*       *OperatorAssociativity*                                                                                            |
| ---------------------------------- | ---------------------------------------------------------------- |----------------------------------------------------------------------------- |
| *Meaning*                        | Decides *which operator is evaluated first* when multiple operators appear in an expression. | Decides *the order (direction)* in which operators of the *same precedence* are evaluated.                        |
| *Purpose*                        | To determine priority among different operators.                                             | To determine direction of execution when priority is the same.                                                      |
| *Example (different precedence)* | 2 + 3 * 4 → * has higher precedence, so → 2 + (3*4) → 14.                              | 10 - 5 + 2 → both - and + have same precedence, so associativity (left → right) decides → (10 - 5) + 2 → 7. |
| *Order of Importance*            | Multiplication, Division, etc. before Addition, Subtraction.                                   | Left-to-right or Right-to-left evaluation among same-level operators.                                                 |
| *Mnemonic*                       | “Who goes first.”                                                                              | “Which direction to go.”    

```python
#STRING
#INDEXING #a=0, p=1, p=2, l=3, e=4; so a=1 it will give p as character
a="apple"
print(a[0]) #INDEXING
print(a[0:3]) #SLICING, In this it will minus 1 from 3 always or any digit in place of 3. Like if we give 0:2, it will give ap.
print(a[0:2])

print(a[-1]) #NEGATIVE INDEXING #a=-5, p=-4, p=-3, l=-2, e=-1 in negative
print(a[-4:-1]) #NEGATIVE SLICING
```

```python
# RANGE=[START:STOP:STEP]
b="This is Python class 3"

print(b[8])
print(b[0:8])
print(b[0:8:2])

#NEGATIVE INDEXING & SLICING
print(b[-1]) 
print(b[-10:-1])
print(b[:]) #takes the whole sentence
print(b[3:]) #means from 0 til the end of sentence or any number
print(b[::-1]) #this gives the reverse of the sentence, but have to add number in step position in range.
```
