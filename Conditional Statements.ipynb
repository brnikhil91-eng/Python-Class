MATCH-CASE

MATCH-VALUE:
     CASE OPTION1:
         ACTION
     CASE OPTION2:
         ACTION
     CASE _:
         DEFAULT

```python
role=input("enter your role")

match role:
    case  "admin":
        print("Full access")
    case "Manager":
        print("Limited access")
    case _: #this is like else
        print("Access to only view")
```

```python
status="W"

match status:
    case "P":
        print("Pending")
    case "E":
        print("Closed")
    case "R":
        print("Resolved")
    case "W":
        print("Waiting")
    case _:
        print("Unknown")
```

```python
city="Delhi"

match city:
    case "Delhi" | "Bangalore":
        print("Metro city")
    case _:
        print("tier 2 city")
```
