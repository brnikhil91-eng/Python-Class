FUNCTIONS-
FUNCTIONS ARE FOR REUSABILITY & ORGANISATION

WITHOUT FUNCTIONS- MESSY CODE
WITH FUNCTIONS- CLEAN, MODULAR, PROFESSIONAL CODE

A REUSABLE BLOCK OF CODE THAT PERFROMS A TASK
INSTEAD OF REWRITING THR LOGIC AGAIN & AGAIN, WE DEFINE IT ONCE AND REUSE IT.

SYNTAX:
def function_name():
    code

```python
#USER DEFINED FUNCTION WITHOUT PARAMETER
def greet(): #use def and give custom function name
    print("Hello user, Hope u r doing great")
```

```python
#POSITIONAL AGRUMENTS - should maintain the position and here order matters
def student(name,age):
    print("Name is:",name)
    print("Age is:",age)

student("Rahul",21)
student(21,"Rahul")
```

RETURN VALUES:-
- SENDS A VALUE BACK TO WHERE THE FUNCTION WAS CALLED
- STOPS THE FUNCTION IMMEDIATELY
- MAKES THE FUNCTION REUSABLE

```python
#PRINT vs RETURN
# PRINT IS TO DISPLAY AND RETURN IS TO GIVE THE VALUE BACK
def add(a,b):
    print(a+b)

result=add(8,6)
print(result)
```

```python
def add(a,b):
    return a+b

result=add(5,5)
print(result)
```

```python
#RETURN SINGLE VALUE
def square(num):
    return num*num

print(square(8))
```

```python
#RETURN MULTIPLE VALUES
def calculate(a,b):
    return a+b,a-b

result=calculate(20,5)
print(result)

sum_val,sub_val=calculate(20,5)
print(sum_val)
print(sub_val)
```

```python
#RETURN WITH CONDITION
def check_even(num):
    if num%2==0:
        return "EVEN"
    else:
        return "ODD"

print(check_even(9))
print(check_even(6))
```

```python
#FUNCTION STIOPS WHEN IT EXECUTES RETURN

def find_first_even(nums):
    for n in nums:
        if n%2==0:
            return n
            print("HEY") #This will not be executed since return has been executed
    return None

print(find_first_even([10,12,13,19]))
print(find_first_even([11,17,13,19]))
```

```python
#RETURN LIST

def get_squares(nums):
    result=[]
    for n in nums:
        result.append(n*n)
    return result

print(get_squares([2,5,9]))
```

```python
#RETURN BOOLEAN - since we have used comparision operator, the output will be true/false
def is_positive(num):
    return num>0

print(is_positive(-5))
print(is_positive(9))
```

```python
#RETURN DICTIONARY
def create_student(name,age):
    return {"name":name,"age":age}

print(create_student("Samiksha",25))
```

DOCSTRING:-
A STRING WRITTEN INSIDE A FUNCTION O EXPLAIN WHAT IT DOES

IT HELPS:-
OTHER DEVELOPERS
FUTURE YOU
INTERVIEWERS
DOCUMENTATION TOOLS

IT USES TRIPLE QUOTES - """    """

```python
def add(a,b):
    """
    Adds two numbers.
    """
    return a+b

print(add(7,7))
print(add.__doc__) #To access the docstring written inside
```

```python
#GOOGLE STYLE DOCSTRING
def add(a,b):
    """
    Add two numbers.

    Args: #Arguments
        a(int):First number
        b(int):Second number
    
    Returns:
    int:Sum of a and b
    """
    return a+b
```

```python
#NUMPY STYLE DOCSTRING
def add(a,b):
    """
    Add two numbers.

    Parameters
    ----------
    a:int
         First number
    b:int
         Second number

    Returns
    -------
    int
        Sum of a and b
    """
    return a+b
```

```python
#ARGS - variable positional arguments
#args collects extra positional arguments into a tuple and it is a non keyword argument

#SUM ALL NUMBERS
def total(*numbers): #(*numbers) - can call how many variables we want and any datatype
    total_sum=0
    for n in numbers:
        total_sum+=n
    return total_sum

print(total(10,20,40))
print(total(50,60))
```

```python
def greet(greeting,*names): #greet - fixed argument, *names - variable argument
    for name in names:
        print(greeting,name)

greet("Good Day", "Ram","Sanju","Carola")
```

```python
def show_types(*args):
    for value in args:
        print(value,"->",type(value))

show_types(15,"Sam",99.6)
```

```python
#kwargs - variable keyword arguments
#can give any name for kwargs like(**two) or (**pot) instead of (**kwargs)
def show_details(**kwargs):
    for key,value in kwargs.items():
        print(key,":",value)

show_details(name="Sam", age=25)
```

```python
def check_email(**kwargs):
    if "email" in kwargs:
        return "Emial Provided"
    return "Email Missing"

print(check_email(name="Rahul", email="Test@gmail.com"))
```

```python
#Update default settings
def settings(**red):
    config={"theme":"Light","Font":12}
    config.update(red)
    return config

print(settings(theme="DARK"))
```

```python
#Combined example
def example(a,*args,**kwargs):
    print("a:",a)
    print("args",args)
    print("kwargs",kwargs)

example(1,2,3,4,name="Manoj",song="Retro")
```

LAMBDA (ANONYMOUS FUNCTION) - SMALL FUNCTION WRITTEN IN ONE LINE.

SYNTAX:
lambda parameters: expression

NO RETURN KEYWORD NEEDED.

USE LAMBDA FOR:-
SHORT LOGIC
INSIDE MAP FILTER
ONE-TIME USE

```python
#NORMAL FUNCTION
def square(x):
    return x*x

#Lambda function (single parameters)
square= lambda x:x*x
print (square(6))

#Lambda function(multiple parameters)
add=lambda a,b:a+b
print(add(5,5))
```

```python
#CHECK EVEN
is_even=lambda x:x%2==0
print(is_even(10))

#GET LAST CHARACTER
last_char=lambda word:word[-1]
print(last_char("Python"))

#CONVERT TO  UPPERCASE
upper=lambda word:word.upper()
print(upper("data analytics"))
```

```python
#MAP() - APPLIES FUNCTION TO EVERY ELEMENT, CAN USE FOR BIG DATA

#NORMAL LOOP
nums=[1,2,3]
squares=[]

for n in nums:
    squares.append(n*n)
print(squares)

#USING MAP()
nums=[1,2,3]
squares=list(map(lambda x:x*x,nums))
print(squares)
```

```python
#ADD 10 TO EACH
nums=[5,10,15]
result=list(map(lambda x:x+10,nums))
print(result)

#UPPERCASE NAMES
names=["amit","sanjana"]
result=list(map(lambda x:x.upper(),names))
print(result)
```

```python
#FILTER FUNCTION - RETURNS ONLY ELEMENTS THAT SATISFY THE CONDITION

#NORMAL LOOP
nums=[1,2,3,4]
evens=[]

for n in nums:
    if n%2==0:
        evens.append(n)
print(evens)        

#USING FILTER()
nums=[1,2,3,4]
evens=list(filter(lambda x:x%2==0,nums))
print(evens)
```

```python
#REMOVE EMPTY STRINGS
data=["Rahul","","Sanketh","","Soujanya"]
result=list(filter(lambda x:x!="",data))
print(result)
```

```python
#REDUCE - USED FOR CUMULATIVE CALCULATIONS [HAVE TO IMPORT FOR THIS]

#Sum of all elements
from functools import reduce
nums=[1,2,3,4]
result=reduce(lambda x,y:x+y,nums)
print(result)
```

```python
#CONCATENATE STRINGS
words=["Data","Science","is","a","subject"]
print(reduce(lambda x,y:x+" "+y,words))
```

```python
#FIND MAX
nums=[10,50,20]
print(reduce(lambda x,y:x if x>y else y,nums))
```

SCOPE - LEGB RULE
SCOPE DETERMINES WHERE THE VARIABLE LIVES

LEGB STANDS FOR:
L- LOCAL
E- ENCLOSING
G- GLOBAL
B- BUILT-IN

```python
#Local scope - only inside
def add():
    a=5
    b=6
    print(a+b)

add()
#Here a,b are local variables and they exist only inside function
```

```python
#global scope - throughout
age=24 

def show():
    print(age)

show()
#here age is the global variable
#Function can read global variable
```

```python
#Enclosing scope - nested functions

def outer():
    x=10
    def inner():
        print(x)
    inner()
outer()

#Here x is enclosing scope variable
#Inner function uses variable from outer function
```

```python
#Built-in scope

x="names"
print(x)
a=len(x)
print(a)
range(5)

len=5
print(len)
a=len(x)  #Now len function won't work because value of len was over-written
print(a)
```

```python
del len #deleting the int object variable
print(len("mango"))
```

```python
#global + local same name

x=100
def show():
    x=50
    print("Inside function",x)

show()
print("Outside functino",x)

#Local variaable does not change or affect global variable
```

```python
#Changing global variable inside function
total=0

def add_amount(amount):
    global total
    total+=amount

add_amount(3500)
add_amount(500)
print(total)
```
