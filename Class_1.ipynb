RULES FOR NAMING VARIABLES:
Variable names can only contain letters, digits and underscores (_).
A variable name cannot start with a digit.
Variable names are case-sensitive (myVar and myvar are different).
Avoid using Python keywords (e.g., if, else, for) as variable names.

```python
#RELATIONAL OPERATORS (only gives in true or false)
a=50
b=5

print(a==b) #checks if it is equal to
print(a<b) #less than
print(a>b) #greater than
print(a!=b) #not equal to
print(a<=b) #less than or equal to
print(a>=b) #greater than or equal to 
```

```python
#LOGICAL OPERATORS
#AND CONDITION:T&T=T, T&F=F, F&F=F, F&T=F 
#OR CONDITION:T&T=T, T&F=T, F&F=F, F&T=T
print((a<b) and (b==a))
print((a<b) or (b==a))
print(not(a<b))
```

```python
#BITWISE OPERATORS
print(5&3) #binary and
print(5|3) #binary or, any one can be true
print(~5) #negation or not, we use -(n+1) and it is called 2's compliment.
print(5^3) #binary xor, if both are true it is false
print(5<<1) #left shift, will shift places, ex:0101 it will be 1010
print(5>>1) #right shift, will drop one digit in the right side of the value.  ex:0101, the last 1 will drop so 2.
```

f-string formatting types: 

variable {x}
expression {x+y}
precision {x:2f}
width {x:8}
alignment {x:^8}
padding {x:*^8}
zero padding {x:05}
percentage {x:2%}
commas {x:,}

```python
print(f"The sum of the numbers is {a+b}") #expression

price=9606.5628
print(f"{price:.2f}") #precision

x=45
print(f"{x:9}") # width, gives spaces and then the value in the 9th place

p="python"
print(f"|{p:<15}|") #left alignment
print(f"|{p:>15}|") #right alignment
print(f"|{p:^15}|") #center alignment

print(f"{45:*<6}") #Padding (left, right and center)
print(f"{45:*>6}")
print(f"{45:*^6}")

print(f"{7:03}") #zero padding
print(f"{125:09}")

value=3.5
print(f"{value:07.2f}") #zero padding + percision

ratio=0.7862
print(f"{ratio:.3%}") #percentage formatting

salary=2259000000
print(f"{salary:,}")  #commas(thousands separator)
print(f"{salary:,.2f}")
```
