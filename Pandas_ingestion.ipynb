PANDAS IS A PYTHON LIBRARY USED FOR:

```python
import pandas as pd
```

Series vs dataframe

series is like a single column of data and it is one dimensional 

```python
s=pd.Series([10,20,30,40])
print(s)
```

```python
#custom index

s=pd.Series([85,90,88],index=["Asha","Ravi","Kiran"])
print(s)
```

```python
#Dataframe - It is a table (rows + columns)

data={
    "Name":["Asha","Ravi","Kiran"],
    "Marks":[85,98,67]
}
df=pd.DataFrame(data)
print(df)

print(df.index)
print(df.columns)
print(df.values)
```
