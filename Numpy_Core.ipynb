PYTHON LIST  VS  NUMPY ARRAY

NUMBERS=[1,2,3,4]
DATA=[1,"hello",3.5]

-CAN STORE DIFFERRENT DATATYPES
-STORED IN SCATTERED MEMORY
-SLOWER FOR LARGER CALCULATIONS

NUMPY ARRAY:
arr=np.array([1,2,3,4])

-STORES SAME DATATYPE
-STORED IN CONTINUOUS MEMOFRY
-MUCH FASTER
-SUPPORTS MATHEMATICAL OPERATIONS

```python
#ndarraay : N-dimensional array
arr=np.array([1,2,3,4]) #multiplies the list
print(arr * 2)

#normal list - repeats the lists
numbers=[1,5,3,2]
print(numbers * 2)
```

```python
#CONVERTING LIST TO ARRAY
arr=np.array([10,20,30])
print(arr)

#2D example
arr1=np.array([[1,2,3],
             [4,5,6]])
print(arr1)
```

```python
#np.zeros - creates array filled with zeros

arr=np.zeros(5)
print(arr)

#Array with 2 rows, 3 columns of zeros
arr1=np.zeros((2,3))
print(arr1)
```

```python
#np.arange - similar to range function
np.arange(0,10,2)

#(start,stop,step)
```

```python
#np.linspace - creates evenely spaced numbers
print(np.linspace(0,10,5))   #0 to 10 only need 5 numbers 

#(start,stop,number of values)

#Logspace - evenly spaced log values
print(np.logspace(1,3,3))
```

```python
#np.ones()

arr=np.ones(5)
print(arr)

np.ones((2,3))
#2 rows, 3 columns

#Use when you need default values of 1
```

```python
#np.full - creates an array filled with any specific value and is useful for initialization.

print(np.full(5,7)) #One dimensional

print(np.full((2,4),8))
#(shape, fill value, [dtype]) - here dtype is optional

print(np.full((2,2),5,float))

print(np.full((3,3),True))

print(np.full((3,2),"NA")) #Used when making missing categorical values
```

```python
#if all 5 students get a bonus marks of 2

marks=np.array([80,75,90,85,88])
bonus=np.full(marks.shape,2)

final_marks=marks+bonus
print(final_marks)
```

```python
#np.eye - creates identity matrix [diagonal=1]

print(np.eye(3))

print(np.diag([1,2,3])) #creates diagonal matrix
```

```python
#np.random - creates an array of random numbers between 0 and 1

print(np.random.rand(4))

print(np.random.rand(2,3))

print(np.random.randint(1,10,5)) #creates 5 random numbers between 1 and 9
```

```python
#random choice
colors=["red","blue","orange","green"]
print(np.random.choice(colors))

#Multiple selection
print(np.random.choice(colors,size=5))
```

BROADCASTING - ALLOWS OPERATIONS BETWEEN ARRAYS OF DIFFERENT SHAPES

BROADCASTING RULES-
TWO ARRAYS ARE CAPABLE IF:

1. THEY HAVE SAME SHAPE
     OR

2. ONE OF THEM IS 1
    OR

3. SMALLER ARRAY CAN STRETCH TO MATCH LARGER

```python
arr=np.array([1,2,3])

print(arr+5)
```

```python
#Broadcasting 1D with 2D

arr=np.array([[1,2,3],
              [4,5,6]])

b=np.array([10,20,30])

print(arr+b)
```

```python
#Broadcasting fails because we need to give 3 numbers in a and 3 numbers in b or 2 in a and 2 in b not different numbers
a=np.array(([1,2,3],
            [4,5,6]))

b=np.array([1,2])

print(a+b)
```

```python
#Column broadcasting

arr=np.array([[1,2,3],
              [4,5,6]])

col=np.array([[10],
              [20]])

print(arr+col)
```

VECTORIZATION - DOING OPERATIONS ON ENTIRE ARRAY AT ONCE INSTEAD OF USING LOOPS

VECTORIZATION MAKES IT:
-FASTER
-CLEANER CODE
-LESS MEMORY USAGE

```python
#Square all numbers
arr=np.array([1,2,3,4])
print(arr**2)
```

```python
a=np.array([1,2,3])
b=np.array([4,5,6])

print(a+b)
```

```python
#Conditional vectorization

arr=np.array([10,20,30,40])
print(arr>28) # (>) is a relational operator and gives true or false
```

```python
#Using condition to modify values

arr[arr>25]=100
print(arr)
```

```python
#using where

np.where(arr>25,1,0)
```

```python
#Mathematical operations

arr=np.array([1,2,3,4])

print(np.sort(arr))
print(np.sqrt(arr))
print(np.log(arr))
print(np.sin(arr))
```

```python
#Broadcasting + vectorization together
#broadcasting handles shape & vectorization handles operation

arr=np.array([[1,2,3],
              [4,5,6]])

mean=np.array([2,3,4])

print(arr-mean)
```
