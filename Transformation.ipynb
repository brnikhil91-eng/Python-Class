DATA TRANSFORMATION :-

.map() - REPLACES VALUES USING A DICTIONARY MAPPING

MAP IN PANDS IS USED TO TRANSFORM OR SUBSTITUTE VALUES IN A SINGLE COLUMN(SERIES) USING A MAPPING RULE.
COMMONLY USED IN DATA CLEANING OR FEATURE ENGINEERING.

SYNTAX:- series.map(mapping)


HERE MAPPING CAN BE ANOTHER:

DICTIONARY
A FUNCTION
ANOTHER SERIES

```python
import pandas as pd
data={"City":
      ["Bangalore","Delhi","Mumbai"]
}
df=pd.DataFrame(data)
df
```

```python
#CONVERT CITY TO REGIONS
region_map={"Bangalore":"South",
            "Delhi": "North",
            "Mumbai":"West"
            }

df["Region"]=df["City"].map(region_map)
df
```

```python
#MAPPING USING A FUNCTION

#.map() - CAN APPLY A FUNCTION TO EACH VALUE

df["city_length"]=df["City"].map(len)
df
```

```python
#MAPPING THROUGH A SERIES
city_series=pd.Series({"Bangalore":"South",
                       "Delhi":"North",
                       "Mumbai":"West"})

df["Region"]=df["City"].map(city_series)
df
```

```python
region_map={"Bangalore":"South",
            "Delhi": "North",
            }

df["Region"]=df["City"].map(region_map)
df
```

```python
#FILLING MISSING VALUES USING MAP

df["Region"]=df["City"].map(region_map).fillna("other")
df
```

```python
import pandas as pd
import numpy as np

data={"Customer":["Asha","Bala","Charlie"],
      "Sales":[50000,30000,15000]}

df=pd.DataFrame(data)
```

```python
#APPLYING CUSTOM FUNCTIONS
#.apply() - LETS US RUN CUSTOM LOGIC ON ROWS OR COLUMNS

#Example - classify sales performance

def sales_category(x):
    if x > 40000:
        return "High"
    elif x > 25000:
        return "Medium"
    else:
        return "Low"
    
#Apply function

df["sales_category"]=df["Sales"].apply(sales_category)
df
```

```python
#CREATE A COLUMN CALLED Double_sales USING lambda WITH APPLY FUNCTION
#Apply on specific column

df["Double_sales"]=df["Sales"].apply(lambda x:x*2)
df
```

```python
#Apply on entire dataframe
df.apply(lambda x:x*2)
```

```python
#Apply with axis

#Axis =0        operate on columns
#Axis =1        operate on rows

df.apply(np.sum,axis=0)
#Calculates the sum cloumn-wise
```

```python
data={
    "A":[10,20,30,40],
    "B":[1,2,3,4]
}

df=pd.DataFrame(data)
df
```

```python
#Row wise calculation or operation
df.apply(lambda row:row["A"] + row["B"],axis=1)
```

```python
data={"Revenue":[50000,30000],
       "Cost":[20000,15000]}

df=pd.DataFrame(data)
```

```python
df["Profit"]=df.apply(lambda row:row["Revenue"]-row["Cost"],axis=1)
df
```

```python
import pandas as pd
import numpy as np

data={"Customer":["Asha","Bala","Charlie"],
      "Sales":["50000","30000","15000"]}

df=pd.DataFrame(data)
df.info()
```

```python
#astype - type conversion
#Example: sometimes numbers might be stored as strings

df["Sales"]=df["Sales"].astype(int)
df.info()

#Converting sales value datatype to numeric
```

```python
data={"Customer":["Asha","Bala","Charlie"],
      "Sales":["50000","30000","missing"]}

df=pd.DataFrame(data)
df.info()
```

```python
#Any missing values will be replaced with NaN - coerce
pd.to_numeric(df["Sales"],errors="coerce")
```
