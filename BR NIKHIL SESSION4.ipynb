Task 1: The "Safe Lookup" Dictionary Engine 

```python
inventory = {
    "sku_01": 50,
    "sku_02": 150,
    "sku_03": 75
}
```

```python
# Update 
inventory.update({"sku_04": 200})
```

```python
# Direct bracket notation
sku_02_value = inventory["sku_02"]
print("Value of sku_02:", sku_02_value)
```

```python
# Safe lookup using .get()
sku_05_value = inventory.get("sku_05", 0)
print("Value of sku_05:", sku_05_value)
```

```python
# Keys and Values
print("All SKU IDs:", list(inventory.keys()))
print("All Stock Counts:", list(inventory.values()))
```

```python
# Dictionary as list of tuples
print("Inventory as tuples:", list(inventory.items()))
```

Task 2: Deep Access in Nested JSON-like Structures 

```python
api_response = {
    "status": "success",
    "data": [
        {"id": 1, "info": {"email": "user1@work.com", "tags": ["admin", "dev"]}},
        {"id": 2, "info": {"email": "user2@work.com", "tags": ["guest"]}}
    ]
}
```

```python
# Extraction
user_1_email = api_response["data"][0]["info"]["email"]
print("User 1 Email:", user_1_email)
```

```python
user_1_secondary_tag = api_response["data"][0]["info"]["tags"][1]
print("User 1 Secondary Tag:", user_1_secondary_tag)
```

```python
last_user_id = api_response["data"][-1]["id"]
print("Last User ID:", last_user_id)
```

```python
# Type observation
print("Type of data:", type(api_response["data"]))
print("Type of info:", type(api_response["data"][0]["info"]))
```

Task 4: The CSV-to-Python Parser (Split & Join) 

```python
csv_data = "Apple,iPhone,1200,Silver"

product_details = csv_data.split(",")

# Modifying price
product_details[2] = "1300"
```

```python
# Reconstructing with semicolon
new_csv = ";".join(product_details)
```

```python
# Header generation
header = "BRAND;MODEL;PRICE;COLOR"
final_output = header + "\n" + new_csv
print(final_output)
```

Task 5: Atomic Dictionary Merging & Popping 

```python
personal_info = {"name": "John", "age": 30}
job_info = {"salary": 5000, "role": "Analyst"}

# Merge 
full_profile = personal_info | job_info
print("Full Profile:", full_profile)
```

```python
# Pop age
removed_age = full_profile.pop("age")
print("Removed Age:", removed_age)
```

```python
personal_info.clear()
print("Length after clear:", len(personal_info))
```
