Task 1: Library Inventory Reader (File Reading)

```python
try:
    count = 0
    with open("books.txt","r") as f:
        for line in f:
            if int(line.strip().split(",")[2]) > 2015:
                count += 1
    print(f"Books after 2015: {count}")

    with open("books.txt","r") as f:
        years = [int(line.strip().split(",")[2]) for line in f]
    print(f"Oldest book year: {min(years)}")

except FileNotFoundError:
    print("Error: books.txt not found.")
```

Task 2: Sales Data Logger (File Writing and Appending )

```python
sale = int(input())

with open("sales_log.txt","a") as f:
    f.write(str(sale) + "\n")

try:
    with open("sales_log.txt","r") as f:
        lines = f.readlines()
        print(f"Total entries: {len(lines)}")
        print(f"Latest entry: {lines[-1].strip()}")
except FileNotFoundError:
    print("Error: sales_log.txt not found.")
```

Task 3: Merging Monthly Reports (Context Managers & Multi-file I/O )

```python

try:
    with open("data1.txt","r") as f1, \
         open("data2.txt","r") as f2, \
         open("sum_data.txt","w") as fout:

        for a, b in zip(f1, f2):
            try:
                fout.write(str(int(a.strip()) + int(b.strip())) + "\n")
            except ValueError:
                print("Error: Non-integer value found.")
                exit()

    with open("sum_data.txt","r") as f:
        print(f.read().strip())

except FileNotFoundError:
    print("Error: One or more data files not found.")
```

Task 4: Robust Transaction Processor (Exception Handling)

```python
balance = 0

try:
    with open("transactions.txt","r") as f:
        for i,line in enumerate(f,1):
            action, amount = line.strip().split()
            try:
                amount = int(amount)
            except ValueError:
                print(f"Line {i}: invalid amount '{amount}'; skipping.")
                continue

            balance += amount if action=="deposit" else -amount

    print(f"Final balance: {balance}")

except FileNotFoundError:
    print("Error: transactions.txt not found.")

finally:
    print("Transaction processing completed.")
```

Task 5: Functional Context Manager for Safe File Operations

```python

from contextlib import contextmanager

@contextmanager
def safe_file_manager(file,mode):
    try:
        f = open(file,mode)
        yield f
    finally:
        f.close()
        print("File closed successfully.")

with safe_file_manager("secure.txt","w") as f:
    f.write("Confidential data")
```
