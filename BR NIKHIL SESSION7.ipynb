Task  1: E-commerce Order Processing System (Function Mastery)

```python
orders = [ 
    {"customer_id": 1, "items": [("laptop", 1200, 1), ("mouse", 25, 2)], "customer_type": "premium"}, 
    {"customer_id": 2, "items": [("phone", 800, 1), ("case", 15, 1)], "customer_type": "regular"}, 
    {"customer_id": 3, "items": [("tablet", 400, 2)], "customer_type": "vip"} 
]

# Configuration for discounts
discounts = {"premium": 0.15, "regular": 0.0, "vip": 0.20}
counts = {"premium": 0, "regular": 0, "vip": 0}

print("Order Processing Results:")

for order in orders:
    ctype = order["customer_type"]
    # Calculate original price: sum of (price * quantity)
    original = sum(item[1] * item[2] for item in order["items"])
    
    # Apply discount
    discount_rate = discounts.get(ctype, 0.0)
    final_price = original * (1 - discount_rate)
    
    # Update counts for summary
    counts[ctype] += 1
    
    print(f"Customer {order['customer_id']} ({ctype}): "
          f"Original: ${original:.2f}, Final: ${final_price:.2f} "
          f"({discount_rate * 100:.1f}% discount)")

print("\nShipping Summary:")
for ctype, count in counts.items():
    print(f"{ctype.capitalize()} customers: {count} orders")
```

Task  2: Financial Analytics Calculator (Modular Design) 

```python
p = 250000
r = 3.5/100/12
n = 30*12

payment = (p*r)/(1-(1+r)**(-n))
interest = payment*n - p

print("Monthly:", round(payment,2))
print("Interest:", round(interest,2))
```

```python
initial = 10000
monthly = 500
r = 7/100/12
months = 20*12

future = initial*(1+r)**months
future += monthly*((1+r)**months-1)/r

print("Final Amount:", round(future,2))
```

```python
portfolio = [("stocks",60),("bonds",30),("cash",10)]
risk = {"stocks":8,"bonds":4,"cash":1}

score = sum(risk[a]*p/100 for a,p in portfolio)
print("Risk Score:", round(score,1))
```

Task  3: Social Media Data Analytics Pipeline

```python
social_media_data = [ 
{"post_id": 1, "user_id": 101, "content": "Amazing sunset today!", "likes": 45, 
"shares": 12, "comments": 8, "timestamp": "2024-01-15", "category": "lifestyle"}, 
{"post_id": 2, "user_id": 102, "content": "New product launch", "likes": 230, 
"shares": 45, "comments": 67, "timestamp": "2024-01-15", "category": "business"}, 
{"post_id": 3, "user_id": 103, "content": "Workout motivation", "likes": 89, 
"shares": 23, "comments": 15, "timestamp": "2024-01-16", "category": "fitness"}, 
{"post_id": 4, "user_id": 101, "content": "Travel tips", "likes": 156, "shares": 
34, "comments": 28, "timestamp": "2024-01-16", "category": "travel"} 
]

def calculate_engagement(post):
    interactions = post["likes"] + post["shares"] + post["comments"]
    return interactions

def analytics_report(data):
    total_posts = len(data)
    
    engagements = [calculate_engagement(p) for p in data]
    avg_engagement = sum(engagements) / total_posts
    
    print("Social Media Analytics Report")
    print("="*30)
    print(f"Total Posts: {total_posts}")
    
    # Trending Content
    sorted_posts = sorted(data, key=lambda x: calculate_engagement(x), reverse=True)
    
    print("\nTrending Content:")
    for i, post in enumerate(sorted_posts,1):
        print(f'{i}. "{post["content"]}" - {calculate_engagement(post)} interactions')
    
analytics_report(social_media_data)
```
