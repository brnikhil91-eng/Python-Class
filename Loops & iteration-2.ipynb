LIST COMPREHENSION-
IT IS A SHORTER WAY TO WRITE A FOR LOOP THAT CREATES A LIST

SYNTAX:
[new_value for item in list]

[new_value for item in list/iterable if condition]

```python
#Square numbers

nums=[1,2,3,4]
squares=[]

for n in nums:
    squares.append(n*n)
print(squares)
```

```python
nums=[1,2,3,4] #Shorter version of code

squares=[n*n for n in nums] #[new_value for item in list]
print(squares)
```

```python
#Convert to lowercase

names=["RAHUL","AMUL","NIKETH"]
names_1=[n.lower() for n in names] #[new_value for item in list]
print(names_1)
```

```python
#In a new list filter only even numbers
nums=[1,2,3,4,5,6,7,8]
even=[n for n in nums if n%2==0] #[new_value for item in list/iterable if condition]
print(even)
```

WHEN TO USE LIST COMPREHENSION:

1. YOU ARE CREATING A LIST
2. YOU ARE TRANSFORMING DATA
3. YOU ARE FILTERING ITEMS
4. LOGIC IS SIMPLE

WHEN NOT TO USE LIST COMPREHENSION:

1. WHEN YOU ARE NOT BUILDING A LIST
2. WHEN LOGIC IS COMPLEX
3. WHEN YOU NEED BREAK OR CONTINUE
4. WHEN UPDATING MULTIPLE VARIABLES

DICTIONARY COMPREHENSION-
USED TO CREATE DICTIONARIES DYNAMICALLY

SYNTAX:
{key:value for item in iterable}

```python
#SQUARE MAPPING
nums=[1,2,3]
squares={n:n*n for n in nums}
print(squares)
```

```python
#PRINTING NAMES ALONG WITH THEIR LENGTH
names=["Nidhi","Ramesh"]
lengths={n:len(n) for n in names}
print(lengths)
```

```python
#Mark students as pass/fail
marks={"Rahul":30,"Anit":60,"Payal":25,"Karan":80}
status={name:"Pass" if score>=40 else "Fail" for name,score in marks.items()}
print(status)
```

```python
#Keys in lowercase & values in uppercase

data={"BOX":"red","HAMMER":"black","SOCKS":"purple"}
new_data={k.lower():v.upper() for k,v in data.items()}
print(new_data)
```

NESTED LOOPS:-
-LOOP INSIDE ANOTHER LOOP
-USED FOR MATRIX,COMBINATIONS, COMPARISONS, GRID LOGIC

SYNTAX:
for x in list1:
    for y in list2:
         action

```python
#Combinations

colors=["Red","Blue"]
sizes=["S","M"]

for c in colors:
    for s in sizes:
        print(c,s)
```

```python
#Pattern printing

for i in range(1,4):
    for j in range(i):
        print("*",end="")
    print()
```

```python
for i in range(1,4):
    for j in range(1,4):
        print(i*j,end=" ")
    print()
```

```python
#Matrix traversal

matrix=[[1,2],[3,4]]

for row in matrix:
    for value in row:
        print(value)
```

```python
#Count elements in matrix
matrix=[[1,2],[3,4]]
count=0

for row in matrix:
    for value in row:
        count+=1
print("Total elements:",count)
```

```python
#Find common names
students=["Rohit","Amit","Priya"]
present=["Amit","Priya"]

for s in students:
    for p in present:
        if s == p: # == means checking if it is true or false
            print(p)
```

```python
#Number triangle

for i in range(1,4):
    for j in range(i):
        print(i,end=" ")
    print()
```

```python
#Square of numbers

for i in range(1,4):        # outer loop - rows
    for j in range(1,5):    # inner loop - numbers 1 to 4
        print(j, end=" ")
    print()
```
