Task 1: Series Fundamentals and Manual DataFrame Creation

```python
import pandas as pd

# Create Series
employee_ages = pd.Series([25, 30, 22, 35, 28],
                          index=['Alice','Bob','Charlie','David','Eve'])

# Print Series
print(employee_ages)

# Print attributes
print("Index:", employee_ages.index)
print("Values:", employee_ages.values)
print("Datatype:", employee_ages.dtype)
```

```python
data = {
    'Name': ['Alice','Bob','Charlie','David','Eve'],
    'Department': ['HR','IT','Finance','HR','IT'],
    'Salary': [60000,75000,62000,80000,70000]
}

employee_data_df = pd.DataFrame(data)

print(employee_data_df)
```

```python
print("Series Shape:", employee_ages.shape)
print("DataFrame Shape:", employee_data_df.shape)

print("David's Age:", employee_ages['David'])

print("Average Age:", employee_ages.mean())
```

Task 2: Basic DataFrame Inspection & Operations (CSV Data)

```python
import pandas as pd

# Load CSV file
fitness_df = pd.read_csv("fitness_data.csv")

# First 5 rows
print(fitness_df.head())
# Last 3 rows
print(fitness_df.tail(3))

# Summary of DataFrame
print(fitness_df.info())

# Descriptive statistics
print(fitness_df.describe())
# Shape (rows, columns)
print("Shape:", fitness_df.shape)

# Data types
print("Data Types:\n", fitness_df.dtypes)

# Column names
print("Columns:", fitness_df.columns)

# Total Calories burned
print("Total Calories:", fitness_df["Calories"].sum())

# Average Duration
print("Average Duration:", fitness_df["Duration"].mean())
```

Task 3: Loading Data from Excel & Attribute Access

```python
import pandas as pd

dummy_excel_data = {
'Month': ['Jan','Feb','Mar','Apr'],
'Year': [2024,2024,2024,2024],
'Enrollments': [150000,165000,148000,172000],
'Updates': [30000,32500,29000,35000],
'Region': ['North','South','East','West']
}

dummy_excel_df = pd.DataFrame(dummy_excel_data)

dummy_excel_df.to_excel('biometric_updates.xlsx', index=False)

print("Created biometric_updates.xlsx")
```

```python
import pandas as pd

biometric_df = pd.read_excel("biometric_updates.xlsx")

print(biometric_df.head(3))
```

```python
print("Shape:", biometric_df.shape)

print("Columns:", biometric_df.columns)

print("Data Types:\n", biometric_df.dtypes)
```

```python
print(biometric_df[['Month','Enrollments']])

print(biometric_df[biometric_df['Enrollments'] > 160000])
```

Task 5: Comprehensive Operations & Real-World Scenario (CSV Data)

```python
import pandas as pd

sales_df = pd.read_csv("product_sales.csv")

print(sales_df.head(7))

print(sales_df.info())
```

```python
sales_df["TotalRevenue"] = sales_df["QuantitySold"] * sales_df["UnitPrice"]

print(sales_df.head())

category_revenue = sales_df.groupby("Category")["TotalRevenue"].sum()

print(category_revenue)

top_products = sales_df.sort_values(by="QuantitySold", ascending=False)

print(top_products.head(5))

electronics_sales = sales_df[sales_df["Category"] == "Electronics"]

print(electronics_sales)
```
