LOOPS

LOOPS ARE USED TO EXECUTE THE SAME BLOCK OF CODE MULTIPLE TIMES WITHOUT WRITING THE CODE.

THERE ARE 2 TTYPES:-
FOR LOOP
WHILE LOOP

FOR LOOP-
RUNS ONCE FOR EVERY ITEM IN A COLLECTION
COLECTION= LIST,STRING,DICTIONARY,etc

SYNTAX-
for item in iterable:
    action

```python
student=["Raj","Sam","Max","Rekha"]

for name in student:
    print(name)
```

```python
genders=["M","F","m"]

for g in genders:
    print(g.lower())
```

```python
#With condition
marks=[30,50,90]

for m in marks:
    if m>=40:
        print("Pass")
    else:
        print("Fail")
```

```python
word="DATA" #every letter as 1 item since it is a string
for w in word:
    print(w)
```

```python
#checking for vowels
word="daTA"
for ch in word:
    if ch.lower() in "aeiou":
        print(ch)
```

```python
words="data"
for c in words:
    if c in "aeiou":
        print(c)
```

```python
emp={"name":"rahul","dept":"IT"}

for l in emp: #Prints only keys
    print(l)

for o in emp.values(): #Prints only values
    print(o)

for l,o in emp.items(): #Prints keys and values both
    print(l,o)
```

```python
#Email validation
emails=["abc@gmail.com","","1235@gmail.com",""]

for e in emails:
    if not e:
        print("Missing")
    else:
        print("Send mail to :",e)
```

FOR -> FIXED COLLECTION
WHILE -> RUN UNTIL CONDITION BEECOMES FALSE

WHILE LOOP-

SYNTAX:
WHILE CONDITION:
     ACTION

```python
#Counting example
i=1

while i<=5:
    print(i)
    i+=1
```

```python
#STOP AT FIRST FAILURE
marks=[70,80,30,90]

for m in marks:
    if m<40:
        print("Fail found")
        break

#break statement is to stop the loop or program at a certain point
#Here break is used to stop at the first failure
```

```python
#CONTINUE
values=["Rahul","","Amit"]

for v in  values:
    if not v:
        continue
    print(v)

#empty string is skipped
#continue is used to skip current iteration and go to next - used to skip 
```

```python
#PASS - placeholder for future code

for n in range(3):
    if n==1:
        pass #do nothing
    else:
        print(n)
```

```python
#Loop through a list using while

names=["Rahul","Amit","Priya"]
i=0
while i<len(names):
    print(names[i])
    i=i+1
```

```python
#Stop when value found and stop the loop when 20 is found

nums=[10,11,12,20,30,40]
i=0     #index position
while i<len(nums):
    if nums[i]==20:
        print("Value found")
        break
    print("number found is:",nums[i])
    i=i+1
```

```python
#Skipping even numbers between 1 - 20

i=0
while i<=20:
    if i%2==0:
        i=i+1
        continue
    print(i)
    i=i+1
```

```python
#Skipping odd numbers between 1 - 20

i=0
while i<=20:
    if i%2!=0: # (!=) means not equal to
        i=i+1
        continue
    print(i)
    i=i+1
```

```python
#Find maximum in a list

nums=[10,45,20,99,30]
max=nums[0]

i=0
while i<len(nums):
    if nums[i]>max:
        max=nums[i]
    i=i+1

print(max)
```

```python
#Print squares upto 10

i=1
while i<=10:
    print(i*i)
    i+=1
```

```python
#Print largest even number

nums=[11,22,33,44,66]
i=0
max=0

while i<len(nums):
    if nums[i]%2==0 and nums[i]>max:
        max=nums[i]
    i+=1
print("Largest even number is:",max)
```
