```python
GUEST = 1       # 0001
USER = 2        # 0010
ADMIN = 4       # 0100
SUPERUSER = 8   # 1000

# Combine permissions
current_access = GUEST | USER

# Elevate admin security
NEW_SECURE_LEVEL = ADMIN << 1

# Check USER permission
has_user = current_access & USER

# Toggle GUEST permission
current_access ^= GUEST

print("Current Access:", current_access, bin(current_access))
print("New Secure Level:", NEW_SECURE_LEVEL, bin(NEW_SECURE_LEVEL))
print("Has USER :", bool(has_user))
```
