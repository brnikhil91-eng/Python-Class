FILE MODES :-

r READ
w WRITE (overwrites file)
a APPEND (adds content)

```python
#Reading specific number of characters

file=open("data.txt","r")
print(file.read(5)) #Reads first 5 characters
file.close()
```

```python
#Using readlines

file=open("data.txt","r")
lines=file.readlines()
print(lines)
file.close()
```

```python
#Appending to a file

file=open("data.txt","a")
file.write("\nhello world")
file.write("\nnew line")
file.close()
```

```python
#Read full content

file=open("data.txt","r")
content=file.read()
print(content)
file.close()
```

```python
#Reading line by line

file=open("data.txt","r")
for line in file:
    print(line)
file.close()

#Better for large files, saves memory
```

WHEN YOU OPEN A FILE PYTHON CREATES:

1. A CONNECTION TO THE FILE
2. A FILE OBJECT IN MEMORY(RAM)
3. A SYSTEM RESOURCE HANDLE

IF YOU DON'T CLOSE IT THAT CONNECTION STAYS OPEN

IF YOU LEAVE IT OPEN :
 
RESOURCE GETS WASTED
SYSTEM MAY SLOW DOWN
YOU MAY NOT BE ABLE TO OPEN NEW FILES

```python
#Checking file mode

file=open("data.txt","r")
print(file.mode)
file.close()
```

```python
#Context manager
#Writing using with

with open("data.txt","w") as f: #won't have to close the file manually if writtn as f
    f.write("\npyton is easy")

#No need for f.close(), the file closes automatically
```

```python
with open("data.txt","r") as f:
    content=f.read()

print(f.closed)
```

```python
#Exceptions (error hanling)
#try-except

try:
    num=int(input("Enter a number"))
except ValueError:
    print("invalid input")
```

ValueError -        WRONG VALUE
TypeError -         WRONG DATA TYPE
NameError -         VARIABLE NOT DEFINED
ZeroDivisionError - DIVIDE BY ZERO ERROR
FileNotFoundError - FILE MISSING
IndexError -        DICTIONARY KEY MISSING
AttributeError -    METHOD DOESN'T EXIST
ImportError -       MODULE CANNOT BE IMPORTED
Exception -         CATCHES ALMOST ALL ERRORS (not recommended for specific error handling)

```python
try:
    print(age)
except NameError:
    print("Variable not defined")
```

```python
try:
    print(10/10)
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
try:
    open ("abc.txt","r")
except:
    print("File not found")
```

```python
try:
    numbers=[1,2,3]
    print(numbers[5])
except IndexError:
    print("Index out of range")
```

```python
try:
    data={"name":"Samual"}
    print(data["age"])
except KeyError:
    print("key not found")
```

```python
#Except is generic, it catches almost all errors
try:
    x=10/0
except Exception:
    print("Something went wrong")

#Not recommended for specific errors
```

```python
#Catching multiple errors
try:
    a=int(input())
    b=int(input())
    print(a/b)
except (ValueError, ZeroDivisionError):
    print("Invalid input")
```

FINALLY - IT IS A BLOCK THAT ALWAYS RUNS WHETHER THERE IS AN ERROR OR NOT.

BASIC SYNTAX:
try:
    risky code
except:
    handle error
finally:
    always runs


WE USE FINALLY TO:
1. CLOSE FILES
2. CLOSE DATABASE CONNECTIONS
3. RELEASE RESOURCES
4. CLEANUP ACTIONS


```python
try:
    print(10/0)
except ZeroDivisionError:
    print("Cannot divide by zero")
finally:
    print("Done")
```

```python
#FILE HANDLING EXAMPLE

file=None

try:
    file=open("tool.txt","r")
    print(file.read())
except FileNotFoundError:
    print("File not found")
finally:
    if file:
        file.close()
        print("file closed")
```

```python
#WITH RETURN STATEMENT
def test():
    try:
        return "try block"
    finally:
        print("Finally block")

test()
   
```

```python
try:
    print("try block")
except:
    print("Error")
else:
    print("No error")
finally:
    print("finally block")
```
