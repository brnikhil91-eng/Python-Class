TASK 4:-

```python
frontend_devs = {"HTML", "CSS", "JS", "React", "Python"}
backend_devs = {"Python", "Java", "NodeJS", "Go", "JS"}


full_stack = frontend_devs | backend_devs
overlap = frontend_devs & backend_devs
specialists = backend_devs - frontend_devs
unique_to_one = frontend_devs ^ backend_devs

print("Full Stack Skills:", full_stack)
print("Common Skills:", overlap)
print("Backend Specialists:", specialists)
print("Unique to One Group:", unique_to_one)
```

```python
float_tuple = (4.5, 2.1, 9.8, 3.3, 2.1)

# Tuple to List
float_list = list(float_tuple)
float_list.sort()
print(float_list)
```

```python
# List to Set
float_set = set(float_list)
float_set.add(2.1)  # duplicate, size will not change
print(float_set)
```

```python
# Set to Tuple
final_tuple = tuple(float_set)
print(final_tuple)
```

```python
print("Sorted List:", float_list)
print("Set:", float_set)
print("Final Tuple:", final_tuple)

print("List ID:", id(float_list))
print("Set ID:", id(float_set))
```
