DICTIONARIES: -
THEY ARE ORDERED, CHANGABLE & DOES NOT ALLOW DUPLICATES(IT SHOULD BE UNIQUE)
IT IS STORED IN {} BRACKETS
VALUEs CAN BE OF ANY DATATYPE BUT KEYS SHOULD BE IMMUTABLE DATATYPE.
1KEY CAN APPEAR ONLY ONCE

```python
#DEFINING A DICTIONARY
a={"apple":"red","orange":"orange","banana":"yellow"}
b={1:"apple",2:"Orange",3:"Banana"}
print(a)
print(b)
```

```python
#ACCESSING VALUES
print(b[2])
print(b.get(4)) #get method- safe method to access values, if there is no 4 in b it will give as none in the execution
print(b.get(4,"No value found")) #get method with custom message
```

```python
#UPDATING VALUES
a["apple"]="Pink" #if it is present it will update or else it will add at the ending
a["grape"]="green"
print(a)

#UPDATING MULTIPLE VALUES
a.update({"apple":"blue","strawberry":"Pink"})
print(a)
```

```python
#POP - TO REMOVE A SPECIFIC VALUE
a.pop("grape")
print(a)
```

```python
#POP ITEM - REMOVES THE LAST VALUE
a.popitem()
print(a)
```

```python
#CLEAR - CLEARs THE ENTIRE DICTIONARY
a.clear()
print(a)
```

```python
#METHODS FOR VIEW:
print(a.keys()) #gives only the keys
print(a.values()) #gives only the values
print(a.items()) #gives both keys & values; each key:value pair is called item
```

```python
#COPY (shallow copy)
c=a.copy()
print(c)
```

```python
#DICT (another way of creating dictionary)
d=dict(a="blue",b="green")
print(d)

e=[1,2,3,4]
f=dict.fromkeys(e,0) #we can only get one values for one key
print(f)
```

```python
#BUILT-IN FUNCTION FOR DICTIONARIES
print(a)
len(a) #len means- how many key:value pairs it has (apple=key & red =value same as follows for all)
```

IN A DATASET STRINGS OR TEXT DATA CAN HAVE ERRORS SO WE HAVE CERTAIN METHODS FOR CLEANING & SORTING DATA.

SOME EXAMPLES OF UNCLEAN STRINGS:
"  BANGALORE"
"  BANGAORE  "
"BANGALORE  "
"MALE/FEMALE"
"JOHN@GMAIL.COM"
"2024-01-15"

TO CLEAN & RESHAPE DATA BEFORE ANALYSIS WE USE STRING  METHODS LIKE -
strip()
replace()
split()
join()
find()

```python
#STRIP()
dept= "   Sales  "
print(dept)
cleaned_dept=dept.strip() #alters string data
print(cleaned_dept)

dept="sales    "
print(dept)
cleaned_dept=dept.lstrip() #removes spaces from left
print(cleaned_dept)
cleaned_dept=dept.rstrip() #removes spaces from right
print(cleaned_dept)

dept="    sales"
print(dept)
cleaned_dept=dept.rstrip() #removes spaces from right
print(cleaned_dept)
cleaned_dept=dept.lstrip() #removes spaces from left
print(cleaned_dept)

price="$400$"
clean_price=price.strip("$") #strip can also remove specific characters 
print(clean_price)

price="$8601$"
print(price.lstrip("$"))
print(price.rstrip("$"))
```

```python
#REPLACE()
status="Not Available"
new_status=status.replace("Not Available", "N/A") #can replace text/strings
print(new_status)
print(status.replace(" ",""))

salary="$3,40,000"
print(salary.replace(",","")) #removing commas using replace

number="+91-951238-848"
print(number.replace("+91-","")) 

email="register.45#gmail.com"
print(email.replace("#","@")) #replacing # with @ using replace

```

```python
#SPLIT()
full_name="Rahul Sharma" #splitting names using split() by giving space inside the upper inverted commas. 
print(full_name.split(" ")) #It identifies the space in the text/string and the  executes.

email="rahul@gamil.com"
print(email.split("@")) #can also split using characters

date="03-12-2028"
print(date.split("-"))

day,month,year=date.split("-") #can also assign values and split
print(year)
print(month)
print(day)
```

```python
#JOIN()
a=["red","blue","orange"]
b=" ".join(a) #we need to use any delimiter to use join()
print(b)

c="-".join(a)
print(c)

nums=["26","15","2035"] #since we are using string method, the numbers must also be in double inverted commas ("")
date="/".join(nums)
print(date)
```

```python
#FIND()
a="data analysis"
print(a.find("analysis")) #find()- finds the index positioning of the assigned value or character(finds index position of space also)
print(a.find(" "))
print(a.find("a"))
print(a.find("z")) #It will give -1 for any character if the the character is not found

first=a.find("a")
print(a.find("a",first+1)) #second occurance of a
```

```python
#EXTRACTING TEXT BETWEEN DELIMITERS
a="abc[Data]/hhsj"
start=a.find("[")
end=a.find("]")
new_a=a[start+1:end]
print(new_a) 
```

```python
a="real"
print(a.upper()) #converting to uppercase
b="YELLOW"
print(b.lower()) #converting to lower case
c="the united states of america"
print(c.capitalize()) #capitalizing the tarting letter
print(c.title()) #converting to a title
d="   YELLOWS"
print(d.strip().lower()) #stripping and converting to lowercase

price="  $1,25,000"
print(price.strip().replace("$","Rs.").replace(",","")) #stripping and replacing 
```

REGEX - (REGULAR EXPRESSIONS)
A WAY TO SEARCH, EXTRACT, VALIDATE OR REPLACE TEXT USING PATTERNS.

EXAMPLES: -
FIND PHONE NUMBERS
FIND EMIALS
REMOVE SYMBOLS
SPLIT MESSY DATA

PYTHON USES THE RE MODULE FOR THIS.
"import re"

MOST USED FUNCTIONS:-
re.search()
re.findall()
re.sub()
re.split()
re.match()

REGEX SYMBOLS-

PATTERN                         MEANING
\d                              digit
\D                             non-digit
\w                            letter/number
\W                          special character
\s                              space

.                              any character
+                              one or more
*                              zero or more
^                                 start
$                                  end 
[abc]                          any of a/b/c

```python
#re.findall() - Returns all matches
import re
text="Order 123,amount 491"

nums=re.findall(r"\d+",text)
print(nums)
```

```python
re.findall(r"\w+","Data 5Science!") #gives only letters or numbers and exculdes any symbols
```

```python
#re.sub - replaces using pattern
phone="+91-98756 54628"
clean_phone=re.sub(r"\D","",phone) #replacing every non-digit with a empty string
print(clean_phone)

text="Hello@#World"
t=re.sub(r"\W"," ",text) #replacing special characters with a space
print(t)

text="DATA     SCIENCE"
t=re.sub(r"\s+"," ",text)
print(t) #replacing with only one space
```

```python
#RE.SPLIT() - FOR MULTIPLE DELIMITERS

text="Python,SQL | Excel"
e=re.split(r"[,|]",text)
print(e)

date="21-12/2023"
d=re.split(r"[-/]",date)
print(d)
```

```python
#RE.SEARCH() - CHECKS IF IT EXISTS

text="my id is 265"
result=re.search(r"\d+",text)
print(result)
```

```python
#RE.MATCH() - TO FIND THE MATCH AT POSITION 1
print(re.match(r"\d","1abc"))
print(re.match(r"\d","abc1"))
```

```python
#EXTRACT EMAIL USERNAME

email="john15@gmail.com"
re.findall(r"^\w+",email) #"^\w"- means start with any word that can be one or more
```

```python
#VALIDATE PINCODE (MUST BE 6 DIGITS)
code="560098"
c=re.match(r"^\d{6}$",code)
print(c)

d=bool(re.match(r"^\d{6}$",code)) #SINCE WE ARE VALIDATING WE NEED A TRUE OR FALSE RESULT. HENCE BOOL FUNCTION IS USED.
print(d)

e=re.findall(r"^\d{6}$",code)
print(e)
```
