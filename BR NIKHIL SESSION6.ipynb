Task 1: The Secure Data Scrubber (String Iteration) 

```python
text = "Alpha_User 101! #Beta$ & Gamma_99"
result = ""

for ch in text:
    if ch.isalpha():
        result += ch
    elif ch == " ":
        pass
        result += ch

print(result)
```

Task 2: Financial Threshold Monitor (List Iteration)

```python
prices = [120.5, 118.2, 122.0, 115.5, 125.0]
first = prices[0]

for i in range(1, len(prices)):
    change = ((prices[i] - first) / first) * 100
    if change < -5:
        print("SELL ALERT at index", i)
```

Task 3: The API Response Validator (Dict Iteration) 

```python
services = {
    "Auth": "200 OK",
    "Cache": "500 Error",
    "Database": "200 OK",
    "Proxy": "404 Not Found"
}

for name, status in services.items():
    if "200" in status:
        print(name, "Service Healthy")
    else:
        print(name, "Service Critical")
        continue
```

Task 8: The "Dirty Data" Integrator  

```python
logs = [22, 25, "TIMEOUT", 28, 0, "ERROR", 30]

total = 0
count = 0

for item in logs:
    if item == "ERROR":
        break
    if item == "TIMEOUT" or item == 0:
        continue
    total += item
    count += 1

if count > 0:
    print("Average:", total / count)
```

Task 9: The Prime Number Architect 

```python
for num in range(2, 51):
    prime = True
    for i in range(2, num):
        if num % i == 0:
            prime = False
            break
    if prime:
        print(num)
```
